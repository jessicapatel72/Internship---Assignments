# Q1

Scrape the details of most viewed videos on YouTube from Wikipedia.            
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos           
You need to find following details:        
A) Rank    
B) Name    
C) Artist      
D) Upload date     
E) Views     

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\web driver\chromedriver.exe")
time.sleep(3)

#url from which you want to scrap data
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(2)
```

```python
#creating an empty list
rank_list = []
name_list = []
artist_list = []
uploading_date_list = []
views_list = []


#scraping rank of the most viewed video
try:
    rank = driver.find_elements(by=By.XPATH, 
                                value="//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in rank:
        rank_list.append(i.text)
except NoSuchElementException:
    rank_list.append('-')
    
time.sleep(1)
#scraping name of the video
try:
    name = driver.find_elements(by=By.XPATH, 
                                value="//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in name:
        name_list.append(i.text)
except NoSuchElementException:
    name_list.append('-')

time.sleep(1)
#scraping artist name
try:
    artist = driver.find_elements(by=By.XPATH, 
                                  value="//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artist:
        artist_list.append(i.text)
except NoSuchElementException:
    artist_list.append('-')
    
time.sleep(1)
#scraping uploading date of the video
try:
    date = driver.find_elements(by=By.XPATH, 
                                value="//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in date:
        uploading_date_list.append(i.text)
except NoSuchElementException:
    uploading_date_list.append('-')
    
time.sleep(1)
#scraping no of views
try:
    views = driver.find_elements(by=By.XPATH, 
                                 value="//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in views:
        views_list.append(i.text)
except NoSuchElementException:
    views_list.append('-')
    
time.sleep(2)    
# Make dataframe
df = pd.DataFrame({'Rank':rank_list,
                   'Video Name':name_list,
                   'Artist':artist_list,
                   'Upload Date':uploading_date_list,
                   'Views(billions)':views_list})
df
```

# Q2

Scrape the details team India’s international fixtures from bcci.tv.       
Url = https://www.bcci.tv/              
You need to find following details:           
A) Match title (I.e. 1st ODI)     
B) Series      
C) Place      
D) Date      
E) Time     
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\web driver\chromedriver.exe")
time.sleep(3)

#url from which you want to scrap data
url = "https://www.bcci.tv/"
driver.get(url)
time.sleep(2)

#clicking the fixtures icon
button=driver.find_element(by=By.XPATH, 
                                 value="//ul[@class='nav navbar-nav  align-items-center']/li[2]/a")
driver.get(button.get_attribute("href"))
time.sleep(2)
```

```python
#creating an empty list
match_title_list = []
series_list = []
place_list = []
date_list = []
time_list = []


#scraping title of the match    
try:
    title = driver.find_elements(by=By.XPATH, 
                                 value="//div[@class='fix-place ng-binding ng-scope']/span[1]")
    for i in title:
        match_title_list.append(i.text.replace("-"," "))
except NoSuchElementException:
    match_title_list.append('-')
    
time.sleep(1)
#scraping series of the match
try:
    series = driver.find_elements(by=By.XPATH, 
                                 value="//div[@class='fixture-card-main col-lg-3 col-md-6 col-sm-12 ng-scope']/div/div/h5[2]/span")
    for i in series:
        series_list.append(i.text)
except NoSuchElementException:
    series_list.append('-')

time.sleep(1)
#scraping place of the match
try:
    place = driver.find_elements(by=By.XPATH, 
                                 value="//div[@class='fix-place ng-binding ng-scope']")
    for i in place:
        place_list.append(i.text.split("-")[1])
except NoSuchElementException:
    place_list.append('-')
    
time.sleep(1)
#scraping date of the match
try:
    date = driver.find_elements(by=By.XPATH, 
                                 value="//div[@class='fixture-card-main col-lg-3 col-md-6 col-sm-12 ng-scope']/div/div/div/div[1]/h5")
    for i in date:
        date_list.append(i.text)
except NoSuchElementException:
    date_list.append('-')
    
time.sleep(1)
#scraping time of the match
try:
    match_time = driver.find_elements(by=By.XPATH, 
                                 value="//div[@class='fixture-card-main col-lg-3 col-md-6 col-sm-12 ng-scope']/div/div/div/div[2]/h5")
    for i in match_time:
        time_list.append(i.text)
except NoSuchElementException:
    time_list.append('-')
    
time.sleep(2)    
# Make dataframe
df = pd.DataFrame({'Match Title':match_title_list,
                   'Series':series_list,
                   'Place':place_list,
                   'Date':date_list,
                   'Time':time_list})
df
```

# Q5

Scrape the details of trending repositories on Github.com.     
Url = https://github.com/     
You have to find the following details:     
A) Repository title     
B) Repository description     
C) Contributors count     
D) Language used     
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\web driver\chromedriver.exe")
time.sleep(3)

#url from which you want to scrap data
url = "https://github.com/"
driver.get(url)
time.sleep(2)
```

```python
# clicking the explore button
explore_button=driver.find_element(by=By.XPATH, 
                                 value="//ul[@class='d-lg-flex list-style-none']/li[4]/details/summary")
try:
    explore_button.click()
except:
    driver.get(explore_button.get_attribute('href'))
time.sleep(1)

#finding the trending option
trending_button=driver.find_element(by=By.XPATH, 
                                 value="//ul[@class='d-lg-flex list-style-none']/li[4]/details/div/ul/li[5]/a")
try:
    trending_button.click()
except:
    driver.get(trending_button.get_attribute('href'))
time.sleep(1)
```

```python
#empty lists to store the data   
url_list=[]
repository_title=[]
repository_desc=[]
contributors_count=[]
language_list=[]


#scraping urls of repositories
repo_url=driver.find_elements(by=By.XPATH, 
                                 value="//h1[@class='h3 lh-condensed']/a")
for i in repo_url:
    url_list.append(i.get_attribute('href'))

time.sleep(1) 
#scraping title of repositories 
try:
    title=driver.find_elements(by=By.XPATH, 
                                 value="//h1[@class='h3 lh-condensed']")
    for i in title:
        repository_title.append(i.text)
except NoSuchElementException:
    repository_title.append('-')

time.sleep(1)

for i in url_list:
    driver.get(i)
    time.sleep(1)
    
    #scraping description of repositories
    try:
        description = driver.find_element(by=By.XPATH, 
                                 value="/html/body/div[4]/div/main/div[2]/div/div/div[3]/div[2]/div/div[1]/div/p")
        repository_desc.append(description.text)
    except NoSuchElementException:
        repository_desc.append('-')
    
    #scraping the contributors_count
    try:
        count=driver.find_element(by=By.XPATH, 
                                 value="//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        contributors_count.append(count.text)
    except NoSuchElementException:
        contributors_count.append('-')
    time.sleep(2)
    
     #scraping used languages
    lang=[]
    languages=driver.find_elements(by=By.XPATH, 
                                 value="//li[@class='d-inline']//a//span[1]")
    if languages:    
        for i in languages:
            lang.append(i.text)
    else:
        lang.append('--')
    language_list.append(lang)
    
    
time.sleep(1)   
#Making dataframe
df=pd.DataFrame({'Repository Title':repository_title,
                'Repository Description':repository_desc,
                'Contributors Count':contributors_count,
                'Language Used':language_list})
df
```

```python
#clicking on view all datasets option
view=driver.find_element(by=By.XPATH, value="//a[@href='datasets.php']")
view.click()
time.sleep(1)
```

```python
#empty lists to store the data    
name_list=[]
type_list=[]
task_list=[]
attribute_list=[]
no_of_instance=[]
no_of_attribute=[]
year_list=[]

# Scraping name of the dataset
try:
    name=driver.find_elements(by=By.XPATH, 
                                 value="/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
    for i in name:     
        name_list.append(i.text)
except NoSuchElementException:
    name_list.append('-')

time.sleep(1)
# Scraping datatype
try:
    type=driver.find_elements(by=By.XPATH, 
                                 value="/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
    for i in type[1:]:
        type_list.append(i.text)
except NoSuchElementException:
    type_list.append('-')
    
time.sleep(1)
# Scraping default task
try:
    task=driver.find_elements(by=By.XPATH, 
                                 value="/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]")
    for i in task[1:]:
        task_list.append(i.text)
except NoSuchElementException:
    task_list.append('-')

time.sleep(1)
# Scraping type of attribute
try:
    attribute=driver.find_elements(by=By.XPATH, 
                                 value="/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]")
    for i in attribute[1:]:
        attribute_list.append(i.text)
except NoSuchElementException:
    attribute_list.append('-')

time.sleep(1)
#Scraping no. of instances
try:
    instance=driver.find_elements(by=By.XPATH, 
                                 value="/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]")
    for i in instance[1:]:
        no_of_instance.append(i.text)
except NoSuchElementException:
    no_of_instance.append('-')
    
time.sleep(1)       
# Scraping no. of attributes
try:
    attribute=driver.find_elements(by=By.XPATH, 
                                 value="/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]")
    for i in attribute[1:]:
        no_of_attribute.append(i.text)
except NoSuchElementException:
    no_of_attribute.append('-')
    
# Scrapping the year details
try:
    year=driver.find_elements(by=By.XPATH, 
                                 value="/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]")
    for i in year[1:]:
        year_list.append(i.text)
except NoSuchElementException:
    year_list.append('-')
    
# Creating the dataframe
df=pd.DataFrame({"Dataset Name":name_list,
                 "Data Type":type_list,
                 "Default Task":task_list,
                 "Attribute Type":attribute_list, 
                 "No of Instances":no_of_instance,
                 "No of Attributes":no_of_attribute,
                 "Year":year_list})
df
```
