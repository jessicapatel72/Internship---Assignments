# Q1 - Write a python program to display all the header tags from wikipedia.org

```python
#creating empty list
list = []                                 

#creating python function which take URL as a parameter
def display_header_tags(url):
    
    #getting reponse from specified URL
    page = requests.get(url)

    #getting page content with BeautifulSoup
    soup = BeautifulSoup(page.content)

    #all heading tags in a list
    heading_tags = ["h1", "h2", "h3", "h4", "h5", "h6"]
    
    for tags in soup.find_all(heading_tags):
        
        #assign tags name and text of particular tag to the variable
        tag_name = tags.name
        tag_text = tags.text.strip()
    
        #making a list
        data = {"Tag": tag_name,
            "Text of tag": tag_text}
        list.append(data)
        
    #making a dataframe
    df = pd.DataFrame(list, columns = ['Tag', 'Text of tag'])
    
    return df
```

```python
#calling a function and passing a given URL
display_header_tags("https://en.wikipedia.org")
```

# Q2 - Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

```python
#importing required library
import re

#creating empty list
movie_title = []
year_of_release = []
ratings = []

#creating python function which take URL as a parameter
def display_top_rated_movie_data(url):
    
    #getting reponse from specified URL
    page = requests.get(url)

    #getting page content with BeautifulSoup
    soup = BeautifulSoup(page.content)
    
    #Scraping data for movie title
    for i in soup.find_all('td',class_="titleColumn"):
        movie_title.append("".join(re.findall("[a-zA-Z]+", i.text)))

    
    #Scraping data for movie's release year 
    for j in soup.find_all('span',class_="secondaryInfo"):
        year_of_release.append(j.text.strip("()"))
        
    
    #Scraping data for ratings of the movie
    for k in soup.find_all('td',class_="ratingColumn imdbRating"):
        ratings.append(k.text.replace("\n",""))
        
        
    #Making a dataframe
    product_data = pd.DataFrame({'Movie Title':movie_title,'Year_of_Release':year_of_release,'Ratings':ratings})
    
    return product_data[0:100]
    
```

```python
#calling a function and passing a given URL
display_top_rated_movie_data("http://www.imdb.com/chart/top")
```

# Q5 - Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
#creating empty list
team = []  
ratings = []
full_info = []
matches = []
points = []

#creating python function which take URL as a parameter
def display_odi_team_details(url):
    
    #getting reponse from specified URL
    page = requests.get(url)

    #getting page content with BeautifulSoup
    soup = BeautifulSoup(page.content)
    
    #Scraping data for team name
    for i in soup.find_all('span',class_="u-hide-phablet"):
        team.append(i.text)
    
    #Scraping data for ratings of a team
    for j in soup.find_all('td',class_="rankings-block__banner--rating u-text-right"):
        ratings.append(j.text.strip())

    for j in soup.find_all('td',class_="table-body__cell u-text-right rating"):
        ratings.append(j.text)
    
    #Scraping data for mathces and points of a team
    for k in soup.find_all('td',class_="table-body__cell u-center-text"):
        full_info.append(k.text)
        
    #matches
    for m in soup.find_all('td',class_="rankings-block__banner--matches"):
        matches.append(m.text)
    
    for m in range(0,len(full_info),2):
        matches.append(full_info[m])
        
    #points
    for n in soup.find_all('td',class_="rankings-block__banner--points"):
        points.append(n.text)

    for n in range(1,len(full_info),2):
        points.append(full_info[n])
    
    #Making a dataframe
    team_details = pd.DataFrame({})
    team_details['Team'] = team[0:10]
    team_details['Matches'] = matches[0:10]
    team_details['Points'] = points[0:10]
    team_details['Ratings'] = ratings[0:10]
    
    return team_details
    
```

```python
#calling a function and passing a given URL for Rajaji Nagar location
display_odi_team_details("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

# Q7 - Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
#creating empty list
heading_name = []  
date = []
content = []
video_link = []

#creating python function which take URL as a parameter
def display_post_details(url):
    
    #getting reponse from specified URL
    page = requests.get(url)

    #getting page content with BeautifulSoup
    soup = BeautifulSoup(page.content)
    
    #Scraping data for heading name of a post
    for i in soup.find_all('a',class_="entry-title-link"):
        heading_name.append(i.text)
        
    #Scraping data for posting date
    for j in soup.find_all('time',class_="entry-time"):
        date.append(j.text)
    
    #Scraping data for content of a post
    for k in soup.find_all('div',class_="entry-content"):
        content.append(k.text.replace("\n",""))
    
    #Scraping data for video link 
    for m in soup.find_all('iframe',class_="youtube-player"):
        video_link.append(m['src'])

    video_link.insert(4,"NA")   #Here at 4th index, there is no any video link so need to add something to make dataframe
    
    #making a dataframe
    post_details = pd.DataFrame({'Heading_name':heading_name,'Date':date,'Content':content,'Video_link':video_link})

    return post_details
```

```python
#calling a function and passing a given URL
display_post_details("https://coreyms.com/")
```

# Q8 - Write a python program to scrape house details from mentioned URL. It should include house title, location, area, EMI and price from https://www.nobroker.in/ .Enter three localities which are Indira Nagar, Jayanagar, Rajaji Nagar.

```python
#creating empty list
house_title = []  
location = []
full_info = []
area = []
emi = []
price = []

#creating python function which take URL as a parameter
def display_house_details(url):
    
    #getting reponse from specified URL
    page = requests.get(url)

    #getting page content with BeautifulSoup
    soup = BeautifulSoup(page.content)
 
    #Scraping data for house title
    for i in soup.find_all('span',class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full"):
        house_title.append(i.text)

    #Scraping data for location of a house
    for j in soup.find_all('div',class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0 po:max-w-95"):
        location.append(j.text)
        
    #Scraping data for area, emi and price of a house
    for k in soup.find_all('div',class_="font-semi-bold heading-6"):
        full_info.append(k.text)
        
    #area
    for x in range(0,len(full_info),3):
        area.append(full_info[x])
    
    #emi
    for y in range(1,len(full_info),3):
        emi.append(full_info[y])
    
    #price
    for z in range(2,len(full_info),3):
        price.append(full_info[z])
    
    
    #makng a dataframe
    house_details = pd.DataFrame({"House_title":house_title,"Location":location,"Area":area,"EMI":emi,"Price":price})

    return house_details
```

# Q9 - Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL

```python
#creating empty list
restaurant_name = []  
location = []
cuisine = []
ratings = []
image_link = []

#creating python function which take URL as a parameter
def display_restaurant_details(url):
    
    #getting reponse from specified URL
    page = requests.get(url)

    #getting page content with BeautifulSoup
    soup = BeautifulSoup(page.content)
    
    #Scraping data for restaurant name
    for i in soup.find_all('a',class_="restnt-name ellipsis"):
        restaurant_name.append(i.text)

    #Scraping data for location of a restaurant
    for j in soup.find_all('div',class_="restnt-loc ellipsis"):
        location.append(j.text)
    
    #Scraping data for cuisine of a restaurant
    for k in soup.find_all('span',class_="double-line-ellipsis"):
        cuisine.append(k.text.split("|")[1])
    
    #Scraping data for ratings of a restaurant
    for m in soup.find_all('div',class_="restnt-rating rating-4"):
        ratings.append(m.text)
        
    #Scraping data for image-link of a restaurant image
    for n in soup.find_all('img',class_="no-img"):
        image_link.append(n['data-src'])
        

    #making a dataframe
    restaurant_details = pd.DataFrame({'Restaurant_name':restaurant_name,'Location':location,'Cuisine':cuisine,
                                       'Ratings':ratings, 'Image_link':image_link})

    return restaurant_details
```

```python
#calling a function and passing a given URL
display_restaurant_details("https://www.dineout.co.in/delhi-restaurants/buffet-special")
```
